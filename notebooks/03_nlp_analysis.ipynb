```python
# Imports
import pandas as pd
import numpy as np
import torch
from IPython.display import display, HTML
from torch.utils.data import Dataset, DataLoader
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments, pipeline, TrainingArguments, DataCollatorWithPadding 
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, precision_recall_curve, classification_report
import nltk
import matplotlib.pyplot as plt
from tqdm import tqdm
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
import json
import spacy
```

```python
with open("../data/processed/processed_songs.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)
#print(df.head())
```

```python
# https://huggingface.co/datasets/ofai/GerMS-AT
dataset = load_dataset("ofai/GerMS-AT")
```

```python
# Labels vorbereiten
def preprocess_labels(example):
    labels_raw = [ann['label'] for ann in example['annotations']]
    numeric_labels = [int(str(l)[0]) for l in labels_raw]  # nur Zahl extrahieren
    example['label'] = int(round(np.mean(numeric_labels)))  # 0-4
    return example

dataset = dataset.map(preprocess_labels)
```

```python
# Binary Labels für Auswertung
def add_binary_label(example):
    example['binary_label'] = 0 if example['label'] == 0 else 1
    return example

dataset = dataset.map(add_binary_label)
```

```python
# Sicherstellen, dass alles Strings sind und keine Nones
train_texts = [str(x) for x in dataset['train']['text'] if x is not None]
test_texts  = [str(x) for x in dataset['test']['text']  if x is not None]
```

```python
# https://huggingface.co/ankekat1000/toxic-bert-german?utm_source=chatgpt.com
# === 4. Tokenizer und Modell laden ===
model_name = "ankekat1000/toxic-bert-german"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
```

```python
def tokenize_fn(texts):
    return tokenizer(texts, padding=True, truncation=True, max_length=256)

train_encodings = tokenize_fn(train_texts)
test_encodings = tokenize_fn(test_texts)
```

```python
# Labels
train_labels = [example['binary_label'] for example in dataset['train']]
test_labels  = [example['binary_label'] for example in dataset['test']]
```

```python
# Dann Binary Labels für Toxic-BERT
train_labels_binary = [0 if l == 0 else 1 for l in train_labels]
test_labels_binary  = [0 if l == 0 else 1 for l in test_labels]
```

```python
class SexismDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels
    def __len__(self):
        return len(self.labels)
    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item
```

```python
# Dataset mit den Binary Labels erstellen
train_dataset = SexismDataset(train_encodings, train_labels_binary)
test_dataset  = SexismDataset(test_encodings, test_labels_binary)
```

```python
# === 7. Training vorbereiten ===
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    logging_dir='./logs',
    learning_rate=2e-5,
    save_strategy='epoch', 
    eval_strategy='epoch',
    load_best_model_at_end=True,
    save_total_limit=2,
    logging_steps=10,
)
```

```python
# Data collator für dynamisches Padding
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
# === Metrics ===
def compute_metrics(pred):
    labels = pred.label_ids
    preds = np.argmax(pred.predictions, axis=-1)
    
    # 0-4 Metriken
    acc = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds, average='macro')
    precision = precision_score(labels, preds, average='macro', zero_division=0)
    recall = recall_score(labels, preds, average='macro', zero_division=0)
    
    # Binary Metriken
    binary_labels = (labels != 0).astype(int)
    binary_preds = (preds != 0).astype(int)
    binary_acc = accuracy_score(binary_labels, binary_preds)
    binary_f1 = f1_score(binary_labels, binary_preds, zero_division=0)
    binary_precision = precision_score(binary_labels, binary_preds, zero_division=0)
    binary_recall = recall_score(binary_labels, binary_preds, zero_division=0)
    
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall,
        'binary_accuracy': binary_acc,
        'binary_f1': binary_f1,
        'binary_precision': binary_precision,
        'binary_recall': binary_recall
    }
```

```python
# === Trainer initialisieren ===
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    processing_class=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)
```

```python
# === 8. Training starten ===
trainer.train()
```

```python
# F1-Scores für alle Thresholds
f1_scores = 2 * (precisions[:-1] * recalls[:-1]) / (precisions[:-1] + recalls[:-1] + 1e-8)
```

```python
# Bestes F1
best_idx_f1 = np.argmax(f1_scores)
best_threshold_f1 = thresholds[best_idx_f1]
```

```python
# Funktion um Metriken zu berechnen
def compute_metrics(threshold, probs, labels):
    preds = (probs >= threshold).astype(int)
    return {
        "Threshold": threshold,
        "Accuracy": accuracy_score(labels, preds),
        "Macro F1": f1_score(labels, preds, average='macro'),
        "Weighted F1": f1_score(labels, preds, average='weighted'),
        "Precision": precision_score(labels, preds),
        "Recall": recall_score(labels, preds),
    }
```

```python
# Metriken für optimierte Thresholds
metrics_list = [
    compute_metrics(best_threshold_f1, probs_test, labels_test),
    compute_metrics(best_threshold_recall, probs_test, labels_test),
    compute_metrics(best_threshold_precision, probs_test, labels_test)
]
```

# BERT ANALYSE

```python
# Aggregation nach Künstlern
alpha = 2  # Smoothing-Faktor
K = 2      # Klassen: sexistisch / nicht-sexistisch

artist_grouped = df.groupby('artist_name').agg(
    avg_sexist_prob=('sexist_prob', 'mean'),
    num_songs_total=('lyrics_clean', 'count'),
    num_songs_sexist=('is_sexist', 'sum')
).reset_index()
```

```python
# Gewichtete Prozentzahl berechnen (Smoothing)
artist_grouped['pct_songs_sexist_weighted'] = (
    (artist_grouped['num_songs_sexist'] + alpha) / 
    (artist_grouped['num_songs_total'] + alpha * K) * 100
)
```

```python
# Echte Quote ohne Smoothing
artist_grouped['pct_songs_sexist'] = (
    artist_grouped['num_songs_sexist'] / artist_grouped['num_songs_total'] * 100
)
```

```python
# Kleine Rundung, um Float-Abweichungen zu vermeiden
artist_grouped['pct_songs_sexist_weighted_round'] = artist_grouped['pct_songs_sexist_weighted'].round(2)
```

```python
# Sortieren nach:
# 1. avg_sexist_prob absteigend
# 2. pct_songs_sexist_weighted absteigend
# 3. num_songs_sexist absteigend
artist_grouped_sorted_bert = artist_grouped.sort_values(
    by=['avg_sexist_prob', 'pct_songs_sexist_weighted_round', 'num_songs_sexist'],
    ascending=[False, False, False]
).reset_index(drop=True)
```

- Abwertungsgrad: Wie beleidigend oder herabwürdigend der Begriff für eine Person ist.
    - Z. B. „Zicke“ → mild, „Kackfotzenhurengesicht“ → extrem stark.
- Zielperson: Bezieht sich der Begriff auf Frauen, Männer, Gruppen, neutrale Kontexte? Begriffe, die gezielt Frauen sexualisieren, sind höher bewertet.
- Aggressivität / Schwere Beleidigung: Wörter, die nicht nur sexualisieren, sondern auch extrem beleidigen („Mördermuschi“) bekommen sehr hohe Werte.
- Numerische Skala 0.0–1.0:
    - 0.0 = Eher neutral, Umgangssprache oder mild abwertend, wenig sexualisiert (z. B. „Muddi“, „Mofgirl“)

```python
# Funktion: Wörter aus dem Lexikon im Song finden und gewichtete Summe berechnen
def analyze_lexicon(lyrics):
    words = lyrics.lower().split()
    lex_words = [word for word in words if word in lexicon]
    total_weight = sum(lexicon[word] for word in lex_words)
    return pd.Series([lex_words, len(lex_words), total_weight])
```

```python
# Anwenden auf Lyrics
df[['lex_words_in_song', 'num_lex_words', 'weighted_score']] = df['lyrics_clean'].apply(analyze_lexicon)
```
